# Notebook 1: Data Exploration
Inspect domain dataset quality before training.

```python
import json
import pandas as pd
from pathlib import Path
from collections import Counter
import matplotlib.pyplot as plt

DATA_DIR = Path('../data/processed')
```

```python
# Load train split
records = []
with open(DATA_DIR / 'train.jsonl') as f:
    for line in f:
        records.append(json.loads(line))

df = pd.DataFrame(records)
print(f'Total training samples: {len(df)}')
df.head()
```

```python
# Task type distribution
df['task_type'].value_counts().plot(kind='bar', title='Task Type Distribution')
plt.tight_layout()
plt.show()
```

```python
# Text length distribution
df['text_len'] = df['text'].str.len()
df['text_len'].hist(bins=50, figsize=(10, 4))
plt.title('Text length distribution (characters)')
plt.xlabel('Characters')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print(df['text_len'].describe())
```

```python
# Source distribution
df['source'].value_counts()
```

```python
# Sample a few examples
for _, row in df.sample(3).iterrows():
    print('─' * 70)
    print(f"INSTRUCTION: {row['instruction']}")
    print(f"\nINPUT (first 200 chars): {str(row['input'])[:200]}")
    print(f"\nOUTPUT (first 200 chars): {str(row['output'])[:200]}")
print('─' * 70)
```
